```python
import sys
import time
import random
from random import randint


 
global position
position = -1
board = ['1', '2', '3', '4', '5', '6', '7', '8', '9']
wins = ((0,1,2), (3,4,5), (6,7,8),
        (0,3,6), (1,4,7), (2,5,8),
        (0,4,8), (2,4,6))

# printing the initial board showing positions 
def initial_board():
    print('Welcome to tic tac game')
    print('''
             [1] | [2] | [3] 
             ---------------
             [4] | [5] | [6]
             ---------------
             [7] | [8] | [9] 
        ''')
        
 # printing board  
def print_board():
    dummy = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    for x, i in enumerate(board):
        if i == 'X' or i == 'O':
            dummy[x] = i
    print(' _'*3)
    for x in(0,3,6):
        print('|' + dummy[x] + '|' + dummy[x+1] + '|' + dummy[x+2] + '|')
        print(' -+-+-')
def score(board):
    for w in wins:
        b = board[w[0]]
        if b in 'XO' and all (board[i] == b for i in w):
            return b, [i+1 for i in w]
    return None
 
def over():
    return all (b in 'XO' for b in board)
 
def available(board):
    return [ b for b in board if b not in 'XO']
 
def computer_turn(symbol, board):
    o = 'O' if symbol =='X' else 'X'
    print("\nComputer thinking...")
    time.sleep(1)
    block = None
    options  = [int(s)-1 for s in available(board)]
    for choice in options:
        brd = board[:]
        brd[choice] = symbol
        if score(brd):
            break
        if block is None:
            brd[choice] = o
            if score(brd):
                block = choice
    else:
        choice = block if block is not None else random.choice(options)
    board[choice] = symbol
    return choice+1

def random_computer_turn(symbol, board):
    o = 'O' if symbol =='X' else 'X'
    print("\nComputer thinking...")
    time.sleep(1)
    block = None
    options  = [int(s)-1 for s in available(board)]
    for choice in options:
        brd = board[::]
        brd[choice] = symbol
        if score(brd):
            break
        if block is None:
            brd[choice] = o
            if score(brd):
                block = choice
    else:
        choice = block if block is not None else random.choice(options)
    board[choice] = symbol
    return random.randint(1, 9)

def player_turn(symbol, board):
    options = available(board)
    while True:
        timeout = 2
        print("\nYOUR TURN\nPlace {0}. Positions remaining {1}  ".format(symbol, ' '.join(options)))
        position = input() 
        if str(position) in options:
            break
        print( "Wrong input." )
    board[int(position)-1] = symbol
    return position
 
def computer1(symbol):
    print_board()
    print('\nComputer1 selects {}'.format(computer_turn(symbol, board)))
    return score(board)

def computer2(symbol):
    print_board()
    print('\nComputer2 selects {}'.format(computer_turn(symbol, board)))
    return score(board)
 
def player(symbol):
    print_board()
    print('\nYou selected {}'.format(player_turn(symbol, board)))
    return score(board)

def first_player(symbol):
    print_board()
    print('Player1 selects {}'.format(player_turn(symbol, board)))
    return score(board)

def second_player(symbol):
    print_board()
    print('Player2 selects {}'.format(player_turn(symbol, board)))
    return score(board)

def random_computer(symbol):
    print_board()
    print('\nComputer selects {}'.format(random_computer_turn(symbol, board)))
    return score(board)


print(__doc__)
initial_board()
choice = input('\nSelect one among the following: \n1. Human VS Human \n2. Human VS random computer \n3. human VS smart computer \n4. Smart computer VS Smart computer')
choice = int(choice)
if choice == 1:
    while not over():
        s = first_player('X')
        if s:
            print_board()
            print("\n*********Congratulations!!********\n  Player {0} wins along positions: {1}" .format(s[0], s[1]))
            break
        if not over():
            s = second_player('O')
            
            if s:
                print_board()
                print("\n*********Congratulations!!********\n  Player {0} wins along positions: {1}" .format(s[0], s[1]))
                break
    else:
        print('\nReally close. The game ended in a draw. :(')


elif choice == 2:
    choice = input("\n'X' or 'O' ?  (Use capitals) ")
    while not over():
        if choice is 'X':
            s = random_computer('O')
        else:
            s = random_computer('X')
        if s:
            print_board()
            print("\nYOU LOSE :(\nComputer wins along positions: {} \n  Better Luck next time".format(s[1]))
            break
        if not over():
            s = player(choice)
            if s:
                print_board()
                print("\n*********Congratulations!!********\n  YOU win along positions: {}".format(s[1]))
                break
    else:
        print('\nReally close. The game ended in a draw. :(')
        
    

                
elif choice == 3:
    choice = input("\n'X' or 'O' ?  (Use capitals) ")
    while not over():
        if choice is 'X':
            s = computer1('O')
        else:
            s = computer1('X')
        if s:
            print_board()
            print("\nYOU LOSE :(\nComputer wins along positions: {} \n  Better Luck next time".format(s[1]))
            break
        if not over():
            s = player(choice)
            if s:
                print_board()
                print("\n*********Congratulations!!********\n  YOU win along positions: {}".format(s[1]))
                break
    else:
        print('\nReally close. The game ended in a draw. :(')
        
    

                
elif choice == 4:
    choice = input("\n'X' or 'O' ?  (Use capitals) ")
    while not over():
        if choice is 'X':
            s = computer1('O')
        else:
            s = computer1('X')
        if s:
            print_board()
            print("\n*********Congratulations!!********\n  computer1 win along positions: {}".format(s[1]))
            break
        if not over():
            s = computer2(choice)
            if s:
                print_board()
                print("\n*********Congratulations!!********\n  computer2 win along positions: {}".format(s[1]))
                break

                
    else:
        print('\nReally close. The game ended in a draw. :(')

```
